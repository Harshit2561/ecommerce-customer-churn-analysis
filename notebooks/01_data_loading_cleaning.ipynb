```python
import pandas as pd
import numpy as np

# load data
df = pd.read_csv('../data/raw/online_retail_II.csv')

# view data
df.head()
```

```python
df.shape
df.info()
df.isna().sum()
```

```python
# remove cancelled invoices
df = df[~df['Invoice'].astype(str).str.startswith('C')]

# check shape
df.shape
```

```python
df.head()
```

```python
# remove rows with missing Customer ID
df = df.dropna(subset=['Customer ID'])

# check shape
df.shape
```

```python
# remove invalid quantity and price
df = df[(df['Quantity'] > 0) & (df['Price'] > 0)]

# final shape after Step 3
df.shape
```

```python
# convert InvoiceDate to datetime
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

# convert Customer ID to integer
df['Customer ID'] = df['Customer ID'].astype(int)
```

```python
df.dtypes
```

```python
#adding new coloumn for total amoung
df['TotalAmount'] = df['Quantity'] * df['Price']
```

```python
df[['Quantity', 'Price', 'TotalAmount']].head()
```

```python
df['InvoiceMonth'] = df['InvoiceDate'].dt.to_period('M')
```

```python
df[['InvoiceDate', 'InvoiceMonth']].head()
```

```python
# save cleaned dataset
df.to_csv('../data/cleaned/cleaned_retail.csv', index=False)
```
